## Percentage of success test

```python
block_ids = np.arange(0, 30, 1)

def test_block(all_beams, probs, num_candidates, beam_size):
    optimum = np.random.choice(all_beams, p=probs)
    beam_probs = probs[:num_candidates]
    new_candidates = np.arange(0, num_candidates, 1)
    total = sum(beam_probs)
    new_probs = beam_probs/beam_probs.sum()
    assert (1 - sum(new_probs) <= 0.0000000001) 
    selected = np.random.choice(new_candidates, p=new_probs, size=(beam_size,))
    # print(f"Optimum is {optimum}\n, we selected the following {num_candidates}: {selected}")
    return optimum in selected



```
